### Visualization

```python
import pandas as pd
from itertools import groupby
import pandas_datareader.data as web
import datetime

start = datetime.datetime(2016, 1, 1)
end = datetime.datetime(2017, 1, 27)
df = web.DataReader('F', 'iex', start, end)
df.axes
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(df.axes[0],df.open , color = 'green', linestyle = 'dashdot', label = 'open')
plt.plot(df.axes[0], df.close , color = 'blue', linestyle = 'dashed', label = 'close')
plt.ylabel('price')
plt.legend()

#plt.show()
```

```python
## First Panel
panel_1 = plt.subplot(2,1,1)
plt.plot(df.axes[0],df.open , color = 'green', linestyle = 'dashdot', label = 'open')
panel_1.set_ylim([0,50])

## Second Panel
plt.subplot(2,1,2)
plt.plot(df.axes[0], df.close , color = 'blue', linestyle = 'dashed', label = 'close')
panel_1.set_xlim([ df.axes[0][0] , df.axes[0][100]])
```

```python
plt.figure(figsize = (9,5))
plt.scatter(df.index, df.loc[df.index].open, label = 'open')
plt.scatter(df.index, df.loc[df.index].close, label = 'close')

plt.title('Open and Close scatter plot', fontsize = 14)
plt.xlabel('date', fontsize = 10)
plt.ylabel('price', fontsize = 11)
plt.legend(loc = 'upper right')
```

```python
plt.figure(figsize = (9,5))
plt.style.use('classic')
num_bins = 10

plt.hist(df.close,num_bins, facecolor = 'red', alpha = 0.75)
plt.title('Open and Close scatter plot', fontsize = 14)
plt.xlabel('Opening', fontsize = 10)
plt.ylabel('Frequency', fontsize = 11)
```

```python
plt.figure(figsize = (9,5))
plt.style.use('classic')

plt.boxplot([df.open,df.close])
plt.axis([0,2, 8,14])
plt.xlim([0.5,2.5])
plt.xticks([1,2],['first','second'])
plt.title('a whisker plot', fontsize = 10, loc = "left")
```
