### Visualization

```python
import pandas_datareader.data as web
import datetime

start = datetime.datetime(2016, 1, 1)
end = datetime.datetime(2017, 1, 27)
df = web.DataReader('F', 'iex', start, end)
df.axes
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(df.axes[0],df.open , color = 'green', linestyle = 'dashdot', label = 'open')
plt.plot(df.axes[0], df.close , color = 'blue', linestyle = 'dashed', label = 'close')
plt.ylabel('price')
plt.legend()

#plt.show()
```

```python
## First Panel
panel_1 = plt.subplot(2,1,1)
plt.plot(df.axes[0],df.open , color = 'green', linestyle = 'dashdot', label = 'open')
panel_1.set_ylim([0,50])

## Second Panel
plt.subplot(2,1,2)
plt.plot(df.axes[0], df.close , color = 'blue', linestyle = 'dashed', label = 'close')
panel_1.set_xlim([ df.axes[0][0] , df.axes[0][100]])
```
